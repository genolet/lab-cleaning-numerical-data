### Define a function that differentiates between continuous and discrete variables. 
### Hint: The number of unique values might be useful. 
### Store continuous data into a continuous_df variable and do the same for discrete_df

```python
num = customer_df.select_dtypes("number")
num.nunique()

### I will set the threshold to more than 300
```

```python
def cont_disc(df):
    cont_cols = []
    disc_cols = []
    for col in df.columns:
        if df[col].nunique() > 300:
            cont_cols.append(col)
        elif df[col].nunique() < 300:
            disc_cols.append(col)
    return cont_cols, disc_cols
```

```python
cont_disc(num)
print(cont_cols)
print(disc_cols)
```

```python
df_cont = num[cont_cols]
df_disc = num[disc_cols]
display(df_cont)
display(df_disc)
```

### Did you find outliers? Comment what you will do with them

*There seem to be some (if not extreme) outliers in the customer_lifetime_value and total_claim amount columns. We could try to filter out these extreme values by using percentiles*

```python
iqr = np.nanpercentile(df_cont['total_claim_amount'],75) - np.nanpercentile(df_cont['total_claim_amount'],25)
upper_limit = np.nanpercentile(df_cont['total_claim_amount'],75) + 1.5*iqr
print(upper_limit)

extraordinary_points = df_cont[df_cont['total_claim_amount'] > upper_limit]
len(extraordinary_points)  
```

```python
iqr = np.nanpercentile(df_cont['customer_lifetime_value'],75) - np.nanpercentile(df_cont['customer_lifetime_value'],25)
upper_limit = np.nanpercentile(df_cont['customer_lifetime_value'],75) + 1.5*iqr
print(upper_limit)

extraordinary_points = df_cont[df_cont['customer_lifetime_value'] > upper_limit]
len(extraordinary_points) 
```
